```python
import csv as csv 
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("data/train.csv") 
df_test = pd.read_csv("data/test.csv") 
```

```python
df_train.head(3)
```

```python
df_train.head(3)
```

```python
df_train['AgeFill'] = df_train['Age']
df_test['AgeFill'] = df_test['Age']
```

```python
df_train['Gender'] = df_train['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df_test['Gender'] = df_test['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
```

```python
 df_test.head()
```

```python
median_ages = np.zeros((2,3))
median_ages
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df_train[(df_train['Gender'] == i) & \
                              (df_train['Pclass'] == j+1)]['Age'].dropna().median()
 
median_ages
```

```python
df_train['AgeFill'] = df_train['Age']
```

```python
median_ages = np.zeros((2,3))
median_ages

for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df_test[(df_train['Gender'] == i) & \
                              (df_test['Pclass'] == j+1)]['Age'].dropna().median()
 
median_ages

df_test['AgeFill'] = df_test['Age']
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        df_train.loc[ (df_train.Age.isnull()) & (df_train.Gender == i) & (df_train.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
        
for i in range(0, 2):
    for j in range(0, 3):
        df_test.loc[ (df_test.Age.isnull()) & (df_test.Gender == i) & (df_test.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
```

```python
df_test
```

```python
df_train['AgeIsNull'] = pd.isnull(df_train.Age).astype(int)
df_test['AgeIsNull'] = pd.isnull(df_test.Age).astype(int)
```

```python
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch']
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch']
```

```python
df_train['Age*Class'] = df_train.AgeFill * df_train.Pclass
df_test['Age*Class'] = df_test.AgeFill * df_test.Pclass
```

```python
df_train.dtypes[df_train.dtypes.map(lambda x: x=='object')]
df_test.dtypes[df_test.dtypes.map(lambda x: x=='object')]
```

```python
df_train = df_train.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'PassengerId'], axis=1) 
df_test = df_test.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'PassengerId'], axis=1) 
```

```python
df_train = df_train.drop(['Age'], axis=1)
df_test = df_test.drop(['Age'], axis=1)
```

```python
df_train = df_train.dropna()
df_test = df_test.dropna()
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
train_data = df_train.values
train_data
```

```python
test_data = df_test.values
test_data
```

```python
# Import the random forest package
from sklearn.ensemble import RandomForestClassifier 

# Create the random forest object which will include all the parameters
# for the fit
forest = RandomForestClassifier(n_estimators = 100)

# Fit the training data to the Survived labels and create the decision trees
forest = forest.fit(train_data[0::,1::],train_data[0::,0])

# Take the same decision trees and run it on the test data
output = forest.predict(test_data)
```

```python
seedling=open("results_v2.csv", "wb")
test=open('data/test.csv', 'rU')
forest_Csv = csv.writer(seedling)
test_file_object = csv.reader(test) 
test_file_object.next() # Header control
```

```python
i = -1
for row in test_file_object:
    row.insert(0,output[i].astype(np.uint8))
    forest_Csv.writerow(row)
    i += 1
 
test.close()
seedling.close()
```
